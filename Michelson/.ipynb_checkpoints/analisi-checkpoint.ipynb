```python
#%matplotlib widget
import numpy as np
from scipy.signal import find_peaks
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import glob
import menzalib as mz
```

# Determinazione lunghezza d'onda dei laser

```python
def conteggio_picchi(file,dist=30,prom=0.05):
    I=np.transpose(np.loadtxt(file))[1]
    picchi,_=find_peaks(I,np.average(I),distance=dist,prominence=prom)
    return len(picchi)

def figura(file,save,plot=False,dist=30,prom=0.05,xlab='t[s]',ylab='Intensit√†[U.A.]',titolo=None):
    plt.close()
    t,I=np.transpose(np.loadtxt(file))
    picchi,_=find_peaks(I,np.average(I),distance=dist,prominence=prom)
    for i in range(3):
        if i==2: plt.ylabel(ylab)
        plt.subplot(311+i)
        plt.plot(t,I,linewidth=0.4)
        plt.scatter(t[picchi[1]],I[picchi[1]],s=0.1)
        plt.scatter(t[picchi],I[picchi],s=1)
        plt.xlim(t[int(len(I)*i/3)],t[int(len(I)*(i+1)/3)-1])
    if titolo!=None: plt.title=titolo
    plt.xlabel(xlab)
    plt.savefig(save+'.eps',format='eps')
    if plot==False: plt.close() 
    
def aggiusta(file):
    with open(file,'r') as f:
        text=f.readlines()
    for i in range(len(text)):
        while ' .' in text[i]:
            text[i]=text[i].replace(' .','.')
    with open(file,'w') as f:
        f.writelines(text)
```

```python
files=glob.glob('1_Dati_misura_lunghezza_onda/*.txt')
picchi=[]
for file in files:
    #figura(file,'1_immagini/'+file[29:-4])
    picchi.append(conteggio_picchi(file))
```

```python
lunghezze,num,DeltaS=[],[],[]
for file in files:
    lunghezze.append(int(file[29:32]))
    num.append(int(file[33:34]))
    DeltaS.append(int(file[-7:-4]))
for i in range(len(picchi)): print(files[i][29:-4],picchi[i])
```

```python
def costante(x,c):
    return c

possibilit√†=[532,633,650]
lunghezza_onda,dlunghezza_onda=[],[]
for f in possibilit√†:
    lung=[]
    dlung=[]
    for i in range(len(num)):
        if lunghezze[i]==f:
            lung.append(2000*DeltaS[i]/picchi[i])
            dlung.append(mz.drapp(2000*DeltaS[i],2000,picchi[1],3))
    #print(lung,'\n',dlung)
    if f==possibilit√†[-1]: lung,dlung=lung[:-1],dlung[:-1]
    plt.errorbar(np.arange(len(lung)),lung,dlung,fmt='.')
    a=len(lung)
    lung,dlung=curve_fit(costante,np.arange(len(lung)),lung,sigma=dlung,absolute_sigma=True)
    plt.plot([-0.2,a-0.8],lung[0]*np.array([1,1]))
    plt.plot([-0.2,a-0.8],(lung[0]+np.sqrt(dlung[0][0]))*np.array([1,1]),'--',color='grey')
    plt.plot([-0.2,a-0.8],(lung[0]-np.sqrt(dlung[0][0]))*np.array([1,1]),'--',color='grey')
    plt.xticks([])
    plt.ylabel('Œª [nm]')
    plt.savefig('./2_immagini/'+str(f)+'.eps',format='eps')
    #plt.show()
    lunghezza_onda.append(lung[0])
    dlunghezza_onda.append(dlung[0][0])
    print(lung,dlung)
```
